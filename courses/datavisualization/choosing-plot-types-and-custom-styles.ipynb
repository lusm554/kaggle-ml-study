```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

# Changing styles with seaborn

```python
# path to the file
filepath = '../../data/spotify.csv'

# read file to dataframe
spotify_data = pd.read_csv(filepath, index_col="Date", parse_dates=True)

# line chart
plt.figure(figsize=(12, 6))
sns.lineplot(data=spotify_data)
```

```python
# change the style of figure to the "dark" theme
sns.set_style("dark")
# line chart
plt.figure(figsize=(12, 6))
sns.lineplot(data=spotify_data)
```

```python
def draw(style):
    # change the style of figure to the "dark" theme
    sns.set_style(style)
    # line chart
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=spotify_data)
    
for s in ["darkgrid", "whitegrid", "dark", "white", "ticks"]:
    draw(s)
```
